# Notebook para instalar dependencias y preparar el entorno para ejecutar app.py

Este notebook verifica la instalación de Python, crea un entorno virtual, instala paquetes necesarios y prepara la aplicación `app.py`.

## 1. Verificar e instalar Python y pip

Verifica que Python 3.x y pip estén instalados en el sistema.

```python
import sys, subprocess, shutil
print('Python:', sys.version)
print('pip:', shutil.which('pip'))
print('pip version:')
subprocess.run(['pip', '--version'])
```

## 2. Crear y activar un entorno virtual

Crea un entorno virtual con `python -m venv venv` y actívalo.

```python
import os, venv
venv_dir = 'venv'
if not os.path.isdir(venv_dir):
    venv.EnvBuilder(with_pip=True).create(venv_dir)
print('Entorno virtual creado en', venv_dir)
print('Actívalo con: source venv/bin/activate')
```

## 3. Instalar dependencias desde requirements.txt

Si existe el archivo `requirements.txt`, instálalo.

```python
import os, subprocess
req_file = 'requirements.txt'
if os.path.exists(req_file):
    subprocess.run(['pip', 'install', '-r', req_file])
    print('Instalado:', req_file)
else:
    print('No existe requirements.txt en el proyecto.')
```

## 4. Instalar dependencias manuales faltantes

Instala manualmente los paquetes que no estén en requirements.txt pero que se usan en el código: `Flask`, `pyotp`, `qrcode`, `Pillow`, `pycryptodome`, `requests`.

```python
import subprocess
packages = ['flask', 'pyotp', 'qrcode', 'pillow', 'pycryptodome', 'requests']
subprocess.run(['pip', 'install'] + packages)
print('Instalados los paquetes manuales necesarios.')
```

## 5. Verificar instalación de dependencias opcionales

Comprueba si necesitas instalar herramientas externas adicionales: `pip-audit`, `trivy`, `safety`, `grype`.

```python
import shutil
tools = ['pip-audit', 'safety', 'trivy', 'grype']
for tool in tools:
    path = shutil.which(tool)
    print(f'{tool}:', path or 'NO INSTALADO')
print('Si falta algún scanner externo, instálalo usando apt o la documentación oficial.')
```

## 6. Probar ejecución de app.py

Ejecuta `python app.py` para verificar que Flask arranca y no hay errores de importación.

```python
import subprocess
try:
    result = subprocess.run(['python3', 'app.py'], capture_output=True, text=True, timeout=10)
    print('returncode:', result.returncode)
    print('stdout:
', result.stdout[:2000])
    print('stderr:
', result.stderr[:2000])
except subprocess.TimeoutExpired:
    print('La ejecución tardó demasiado tiempo. Revisa la app en la terminal.')
```
